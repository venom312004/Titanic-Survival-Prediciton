```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("titanic")
df
```

```python
# sex and who both are same so we gonna remove one and also adult_male is useless nd  also embark and embarked town is same also alive and alone
```

```python
df.info()
```

```python
# here deck has many null values so we gonna drop it
```

```python
df.drop(['deck','who','alive','embark_town','adult_male','class'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True) # another method x = df.loc[df['age'].notna(), 'age'].mean() then  df['age'].fillna(x,inplace=True)
```

```python
# in embark either fill null by mean or remove that rows so we gonna remove that row
```

```python
df.dropna(subset=['embarked'],inplace=True) # subset=['embarked'] → only check the embarked column for missing values 
```

```python
df.info()
```

```python
df.head()
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()
df['sex']=model.fit_transform(df['sex'])
df['embarked']=model.fit_transform(df['embarked'])  # C=0,Q=1,S=2 (numbering is based on alphabetical order)
```

```python
df
```

```python
df=df.astype(int)
df
```

```python
# here we not gonna do standard scaling here we just focusing on logistic regression
```

```python
X=df.drop('survived',axis=1)
Y=df['survived']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
Y_test
```

## Logistic Regression

```python
# in classfication problem we use confusion matrix in place of r2 to check our model performance, is it good or bad
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report # in classification_report you'll get everything f1_score,precision_score,recall_score
```

```python
accuracy_score(Y_test,Y_pred)  # syntax and everything is same as r2_score etc 
```

```python
# 80.3 % is our aaccuracy_score
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
# 90 in TP, 53 is TN , 16 is FP and 19 is FN
```

```python
print(classification_report(Y_test,Y_pred))  # F1-score is the mean of precision and recall
```

## KNN

```python
#KNN(K-nearest Neighbour)
# we first gonna do feature scaling bcz in knn distance matter and using standard we close data weights near to predicted one
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
X_train
```

```python
X_train_scale=scale.fit_transform(X_train)
X_train_scale
```

```python
X_test_scale=scale.fit_transform(X_test)
X_test_scale
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model=model.fit(X_train_scale,Y_train)
knn_model
```

```python
Y_pred_knn=knn_model.predict(X_test_scale)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy=accuracy_score(Y_test,Y_pred_knn)
accuracy
```

```python
print(classification_report(Y_test,Y_pred_knn)) # F1-score is the mean of precision and recall
```

```python
confusion_matrix(Y_test,Y_pred_knn)
```

## Support Vector Machine

```python
from sklearn.svm import SVC   # support vector classification
```

```python
model_svm = SVC(kernel='rbf')     # kernel=rbf in hyperplane we need to divide and that's not done by any straight line and they are very close to each other so we use kernel that create hyperplane in n-dimension(3D)
```

```python
model_svm.fit(X_train_scale,Y_train)   # feature scaling need bcz they should be closely to each other
```

```python
Y_pred_SVC=model_svm.predict(X_test_scale)
Y_pred_SVC
```
